```python
import requests
import pandas as pd
import numpy as np
from datetime import date
import matplotlib.pyplot as plt
import os #For using set blocking.
```

```python
df = pd.read_csv('final.csv', index_col=0) #Loading the csv as df.
df
```

```python
stock_list = list(df['stock'])#Check data_frame once ready.
stock_list
```

```python
#Defining the get_quotes function to retrieve ticker information.

def get_quotes(stock_list):
    #Defining the endpoint

    endpoint_quotes = r'https://api.tdameritrade.com/v1/marketdata/quotes'

    #Defining the payload


    payload_quotes = {'apikey': '4JHA7B4AE8VS1DQDUGATGHZPQIZNODEO',
            'symbol': stock_list} #We need the tickers seperated by commas.


    quotes_content = requests.get(url = endpoint_quotes, params=payload_quotes).json()
    
    return quotes_content
```

```python
#Defining the 'get_fundamentals' to retrieve fundamental company information.

def get_fundamental(stock_list):
    
    endpoint_fundamentals = r'https://api.tdameritrade.com/v1/instruments'

    #Defining the payload

    payload_fundamental = {'apikey': '4JHA7B4AE8VS1DQDUGATGHZPQIZNODEO',
                            'symbol': stock_list,
                          'projection':'fundamental'
                          } #We need the tickers seperated by commas.


    fundamental_content = requests.get(url = endpoint_fundamentals, params=payload_fundamental).json()
    
    return fundamental_content
```

```python
#Quotes data_frame

quotes_content = get_quotes(stock_list) #Getting the quotes info from TDAmeri
df_quotes = pd.DataFrame.from_dict(quotes_content, orient = 'index')
df_quotes
df_quotes.reset_index(inplace=True)
# df_quotes.drop('index', axis=1, inplace=True)
# print(df_quotes)
df_quotes
#df_quotes.set_index('stock')
```

```python
#Fundamental data_frame

# Establish columns for fundamental data in the datafrane

fundamental_content = get_fundamental(stock_list)
fund_cols = []
for column in [x for x in [*fundamental_content[stock_list[0]]['fundamental']]]:
     fund_cols.append(column)

# Append data to dataframe
df_fundamental = pd.DataFrame(columns = fund_cols)
for stock in stock_list:
    df_fundamental = df_fundamental.append(pd.Series(fundamental_content[stock]['fundamental'], index=fund_cols), ignore_index=True)

df_temp=df_fundamental
df_temp=df_temp.rename(columns={'symbol':'stock'})
df_temp
df_fundamental = df_temp
df_fundamental
#df_fundamental.set_index('stock')
```

```python
#Merging all three data frames

df_merged = pd.concat([df,df_fundamental], axis=1)

#Removing duplicates
df_merged = df_merged.loc[:, ~df_merged.columns.duplicated()]

```
